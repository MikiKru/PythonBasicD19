```python
#i = 2
oceny2 = oceny.copy()
# 13 zmieniamy
oceny2.rename({"Z13(5)":"Z13(4+1)"}, axis=1, inplace=True)
# 15 usuwamy
oceny2.drop("Z15(5)", axis=1, inplace=True)
oceny2

for i in range(2, 16):
    numer, maks_lab, maks_dom = re.findall("Z(\d{1,2})\((\d+)\+(\d+)\)", oceny2.columns[i])[0]
    #print(numer, maks_lab, maks_dom)
    laboratorium = oceny2.iloc[:, i].str.split("+", expand=True)[0] # expand zwraca każdą składową jako osobną kolumnę
    domowa = oceny2.iloc[:, i].str.split("+", expand=True)[1]
    oceny2[f"L{numer}({maks_lab})"] = laboratorium
    oceny2[f"D{numer}({maks_dom})"] = domowa

oceny2.drop(oceny2.columns[2:16], axis=1, inplace=True)
oceny2
```

# Zamieniamy nieobecności/plagiaty

* NB, NA - nieobecność nieusprawiedliwiona, 0
* US, ZW - nieobecność usprawiedliwiona, o tym zaraz, np.NaN
* NS.(1.0) - praca niesamodzielna, 0

```python
oceny3 = oceny2.copy()
oceny3.replace("NB", "0", inplace=True)
oceny3.replace("NA", "0", inplace=True)
oceny3.replace("NS.*", "0", inplace=True, regex=True)
oceny3.replace("ZW", np.NaN, inplace=True)
oceny3.replace("US", np.NaN, inplace=True)
oceny3
```

# Wstawiamy średnie z średnich za np.NaN (zwolnienia lekarskie)

## Konwersja na floaty

```python
oceny3.iloc[:,3:] = oceny3.iloc[:,3:].astype(float)
oceny3.info()
```

```python
oceny3.columns
```

```python
# wybrać odpowiednie kolumny, skonwertować na floaty
# połączyć concatem imię, nazwisko, e-mail i te skonwertowane kolumny

oceny4 = pd.concat([oceny3.iloc[:,:3],oceny3.iloc[:,3:].astype(float)], axis=1)
oceny4.info()
```

## Pętle

```python
# pętla po wierszach
#    pętla po kolumnach
#      jeśli nan:
#        policz średnią z wiersza i kolumny
```

```python
%%timeit
oceny5 = oceny4.copy()
for wiersz in range(oceny5.shape[0]):
    for kolumna in range(3, oceny5.shape[1]):
        if np.isnan(oceny5.iloc[wiersz, kolumna]):
            srednia_kolumny = oceny5.iloc[:, kolumna].mean()
            srednia_wiersze = oceny5.iloc[wiersz, 3:31].mean()
            oceny5.iloc[wiersz, kolumna] = np.mean([srednia_kolumny, srednia_wiersze])
oceny5
```

### Bardziej zwektoryzowany

```python
# policzyć średnie dla każdej kolumny i każdego wiersza (w sposób zwektoryzowany)
# znaleźć indeksy, na których są nany (zwektoryzowany)
# podstawić pod te indeksy średnie z średnich (indeksowanie listą liczb całkowitych)

oceny_macierz = oceny4.values[:, 3:].astype(float)
oceny_macierz.shape
```

```python
oceny_wiersze = np.nanmean(oceny_macierz, axis=1)
oceny_kolumny = np.nanmean(oceny_macierz, axis=0)
oceny_wiersze
```

```python
indeksy_nan = np.where(np.isnan(oceny_macierz))
indeksy_nan
```

```python
oceny_macierz[indeksy_nan] = (oceny_wiersze[indeksy_nan[0]] + oceny_kolumny[indeksy_nan[1]])/2
```

```python
oceny_macierz
```

```python
oceny5_Numpy = oceny4.copy(deep=True)
oceny5_Numpy.iloc[:, 3:] = oceny_macierz
```

```python
nazwy = oceny5.columns.values.astype(str)
suma_max = np.sum(np.apply_along_axis(lambda x : re.findall("[LD]\d+\((\d+)\)", str(x)),
                           axis=0, 
                           arr=nazwy).astype(int))
suma_max
```

```python
oceny5[f"Suma({suma_max})"] = oceny5.sum(axis=1)
oceny5
```

```python
oceny5[f"Procenty"] = oceny5["Suma(69)"]/suma_max
oceny5
```
